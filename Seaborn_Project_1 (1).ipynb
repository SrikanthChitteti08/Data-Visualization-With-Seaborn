###  **Combining Gender & Marital Status**

```python
lst = []
for i in range(len(df)):
  lst.append(df['Gender'][i] + '_' + str(df['Marital_Status'][i]))
df['Marital_Gender'] = lst
```

### **Total Purchase Amount by Marital Status and Gender**

```python
import seaborn as sns
import matplotlib.pyplot as plt

grouped_data = df.groupby('Marital_Gender')['Purchase'].sum().reset_index()


sorted_data = grouped_data.sort_values(by=['Purchase'], ascending=False)


plt.figure(figsize=(4, 4))
sns.barplot(x='Marital_Gender', y='Purchase', data=sorted_data)


plt.xlabel('Marital Status and Gender')
plt.ylabel('Total Purchase Amount')
plt.title('Total Purchase Amount by Marital Status and Gender')


plt.show()
```
